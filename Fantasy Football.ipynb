## Method to create a dictionary based off of the player name, the upcoming game week, their position, and their team

##### Creates a decision tree, random forest tree, and the associated feature importances and confusion matricies for each.

```python
def testOverperformed(df):
    master_df = df
    print('----------------------------- Overall -----------------------------')
    tree_x = df.drop(['Week', 'Tm','Player','Pos','Status','Actual','Difference','Overperformed'],1)
    tree_y = df.Overperformed
    x_train, x_test, y_train, y_test = train_test_split(tree_x, tree_y, test_size=0.2, random_state=2)
    # model
    tree = DecisionTreeClassifier(random_state=2, min_samples_leaf=5, min_samples_split=20)

    # train
    
    tree.fit(x_train, y_train)
    # score
    print('----------------------------- Decision Tree -----------------------------')
    print('Decision Tree Training Score')
    print(tree.score(x_train, y_train))
    print('Decision Tree Test Score')
    print(tree.score(x_test, y_test))

    gain_df = pd.DataFrame({'Gain': tree.feature_importances_}, index=x_train.columns).sort_values('Gain', ascending = False)
    print('Decision Tree Feature Importances')
    print(gain_df)
    
    print('Decision Tree Confusion Matrix')
    print(pd.crosstab(y_test, tree.predict(x_test), rownames=['Actual'], colnames = ['Predicted:'], margins=True))
    
    print('----------------------------- Random Forest Tree -----------------------------')
    # model
    forest = RandomForestClassifier(random_state=2)
    

    # train
    forest.fit(x_train, y_train)

    # predict
    forest_predictions = forest.predict(x_test)
    print('Random Forest Score')
    print(forest.score(x_test, y_test))

    print('Random Forest Feature Importances')
    importance_df = pd.DataFrame({'Importance': forest.feature_importances_}, index = x_train.columns).sort_values('Importance', ascending = False)
    print(importance_df)

    print('Random Forest Confusion Matrix')
    pd.crosstab(y_test, forest_predictions, rownames=['Actual'], colnames = ['Predicted:'], margins = True)
    
    
    print('----------------------------- Quarterbacks -----------------------------')
    df=master_df[master_df.Pos == 'QB']
    tree_x = df.drop(['Week', 'Tm','Player','Pos','Status','Actual','Difference','Overperformed'],1)
    tree_y = df.Overperformed
    x_train, x_test, y_train, y_test = train_test_split(tree_x, tree_y, test_size=0.2, random_state=2)
    # model
    tree = DecisionTreeClassifier(random_state=2, min_samples_leaf=5, min_samples_split=20)

    # train
    
    tree.fit(x_train, y_train)
    # score
    print('----------------------------- Decision Tree -----------------------------')
    print('Decision Tree Training Score')
    print(tree.score(x_train, y_train))
    print('Decision Tree Test Score')
    print(tree.score(x_test, y_test))

    gain_df = pd.DataFrame({'Gain': tree.feature_importances_}, index=x_train.columns).sort_values('Gain', ascending = False)
    print('Decision Tree Feature Importances')
    print(gain_df)
    
    print('Decision Tree Confusion Matrix')
    print(pd.crosstab(y_test, tree.predict(x_test), rownames=['Actual'], colnames = ['Predicted:'], margins=True))
    
    print('----------------------------- Random Forest Tree -----------------------------')
    # model
    forest = RandomForestClassifier(random_state=2)
    

    # train
    forest.fit(x_train, y_train)

    # predict
    forest_predictions = forest.predict(x_test)
    print('Random Forest Score')
    print(forest.score(x_test, y_test))

    print('Random Forest Feature Importances')
    importance_df = pd.DataFrame({'Importance': forest.feature_importances_}, index = x_train.columns).sort_values('Importance', ascending = False)
    print(importance_df)
    
    print('Random Forest Confusion Matrix')
    pd.crosstab(y_test, forest_predictions, rownames=['Actual'], colnames = ['Predicted:'], margins = True)
    
    
    print('----------------------------- Running Backs -----------------------------')
    df=master_df[master_df.Pos == 'RB']
    tree_x = df.drop(['Week', 'Tm','Player','Pos','Status','Actual','Difference','Overperformed'],1)
    tree_y = df.Overperformed
    x_train, x_test, y_train, y_test = train_test_split(tree_x, tree_y, test_size=0.2, random_state=2)
    # model
    tree = DecisionTreeClassifier(random_state=2, min_samples_leaf=5, min_samples_split=20)

    # train
    tree.fit(x_train, y_train)
    # score
    print('----------------------------- Decision Tree -----------------------------')
    print('Decision Tree Training Score')
    print(tree.score(x_train, y_train))
    print('Decision Tree Test Score')
    print(tree.score(x_test, y_test))

    gain_df = pd.DataFrame({'Gain': tree.feature_importances_}, index=x_train.columns).sort_values('Gain', ascending = False)
    print('Decision Tree Feature Importances')
    print(gain_df)
    
    print('Decision Tree Confusion Matrix')
    print(pd.crosstab(y_test, tree.predict(x_test), rownames=['Actual'], colnames = ['Predicted:'], margins=True))
    
    print('----------------------------- Random Forest Tree -----------------------------')
    # model
    forest = RandomForestClassifier(random_state=2)
    

    # train
    forest.fit(x_train, y_train)

    # predict
    forest_predictions = forest.predict(x_test)
    print('Random Forest Score')
    print(forest.score(x_test, y_test))

    print('Random Forest Feature Importances')
    importance_df = pd.DataFrame({'Importance': forest.feature_importances_}, index = x_train.columns).sort_values('Importance', ascending = False)
    print(importance_df)
    
    print('Random Forest Confusion Matrix')
    pd.crosstab(y_test, forest_predictions, rownames=['Actual'], colnames = ['Predicted:'], margins = True)
    
    
    print('----------------------------- Wide Receivers -----------------------------')
    df=master_df[master_df.Pos == 'WR']
    tree_x = df.drop(['Week', 'Tm','Player','Pos','Status','Actual','Difference','Overperformed'],1)
    tree_y = df.Overperformed
    x_train, x_test, y_train, y_test = train_test_split(tree_x, tree_y, test_size=0.2, random_state=2)
    # model
    tree = DecisionTreeClassifier(random_state=2, min_samples_leaf=5, min_samples_split=20)

    # train
    
    tree.fit(x_train, y_train)
    # score
    print('----------------------------- Decision Tree -----------------------------')
    print('Decision Tree Training Score')
    print(tree.score(x_train, y_train))
    print('Decision Tree Test Score')
    print(tree.score(x_test, y_test))

    gain_df = pd.DataFrame({'Gain': tree.feature_importances_}, index=x_train.columns).sort_values('Gain', ascending = False)
    print('Decision Tree Feature Importances')
    print(gain_df)
    
    print('Decision Tree Confusion Matrix')
    print(pd.crosstab(y_test, tree.predict(x_test), rownames=['Actual'], colnames = ['Predicted:'], margins=True))
    
    print('----------------------------- Random Forest Tree -----------------------------')
    # model
    forest = RandomForestClassifier(random_state=2)
    

    # train
    forest.fit(x_train, y_train)

    # predict
    forest_predictions = forest.predict(x_test)
    print('Random Forest Score')
    print(forest.score(x_test, y_test))

    print('Random Forest Feature Importances')
    importance_df = pd.DataFrame({'Importance': forest.feature_importances_}, index = x_train.columns).sort_values('Importance', ascending = False)
    print(importance_df)
    
    print('Random Forest Confusion Matrix')
    pd.crosstab(y_test, forest_predictions, rownames=['Actual'], colnames = ['Predicted:'], margins = True)
    
    print('----------------------------- Tight Ends -----------------------------')
    df=master_df[master_df.Pos == 'TE']
    tree_x = df.drop(['Week', 'Tm','Player','Pos','Status','Actual','Difference','Overperformed'],1)
    tree_y = df.Overperformed
    x_train, x_test, y_train, y_test = train_test_split(tree_x, tree_y, test_size=0.2, random_state=2)
    # model
    tree = DecisionTreeClassifier(random_state=2, min_samples_leaf=5, min_samples_split=20)

    # train
    tree.fit(x_train, y_train)
    # score
    print('----------------------------- Decision Tree -----------------------------')
    print('Decision Tree Training Score')
    print(tree.score(x_train, y_train))
    print('Decision Tree Test Score')
    print(tree.score(x_test, y_test))

    gain_df = pd.DataFrame({'Gain': tree.feature_importances_}, index=x_train.columns).sort_values('Gain', ascending = False)
    print('Decision Tree Feature Importances')
    print(gain_df)
    
    print('Decision Tree Confusion Matrix')
    print(pd.crosstab(y_test, tree.predict(x_test), rownames=['Actual'], colnames = ['Predicted:'], margins=True))
    
    print('----------------------------- Random Forest Tree -----------------------------')
    # model
    forest = RandomForestClassifier(random_state=2)
    

    # train
    forest.fit(x_train, y_train)

    # predict
    forest_predictions = forest.predict(x_test)
    print('Random Forest Score')
    print(forest.score(x_test, y_test))

    print('Random Forest Feature Importances')
    importance_df = pd.DataFrame({'Importance': forest.feature_importances_}, index = x_train.columns).sort_values('Importance', ascending = False)
    print(importance_df)
    
    print('Random Forest Confusion Matrix')
    pd.crosstab(y_test, forest_predictions, rownames=['Actual'], colnames = ['Predicted:'], margins = True)
```

## Method that takes a defense and a player's position and returns the average points the defense allows to that position

```python
def defAvg(defense, position, week):
    if position == 'HB':
        position = 'RB' #For fantasy purposes these are the same thing
    def_df = result_df[ (result_df.Pos == position) & (result_df.Opp == defense) & (result_df.Week < week)]
    
    
    avg=0
    div = len(def_df.index)
    if def_df.empty:
        return 0
    if week ==1:
        div =1
    avg = def_df.Actual.sum()/div
    return avg
```
